```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler, SplineTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score

# --- 1. Load dataset
df = pd.read_csv("day.csv")

# --- 2. Drop unnecessary columns
df.drop(['instant', 'dteday', 'casual', 'registered'], axis=1, inplace=True)

# --- 3. Define features & target
X = df.drop('cnt', axis=1)
y = df['cnt']

# --- 4. Train/Test split (20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# --- 5. Identify numeric and categorical columns
num_cols = X.select_dtypes(include=np.number).columns.tolist()
cat_cols = X.select_dtypes(exclude=np.number).columns.tolist()

# --- 6. Preprocessing pipelines
numeric_transformer = Pipeline([
    ("scaler", StandardScaler()),
    ("spline", SplineTransformer())  
])

categorical_transformer = Pipeline([
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer([
    ("num", numeric_transformer, num_cols),
    ("cat", categorical_transformer, cat_cols)
])

# --- 7. Create full pipeline with Ridge regression
ridge = Ridge(max_iter=10000)
pipe = Pipeline([
    ("preprocessor", preprocessor),
    ("ridge", ridge)
])

# --- 8. Hyperparameter grid
param_grid = {
    "ridge__alpha": np.logspace(-2, 3, 50),  
    "preprocessor__num__spline__degree": [2, 3, 4], 
    "preprocessor__num__spline__n_knots": [5, 8, 10, 12]  
}

# --- 9. Grid Search CV
grid_search = GridSearchCV(pipe, param_grid, cv=5, scoring="r2", n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)

# --- 10. Best model & prediction on test
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# --- 11. Metrics on test set
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Best params: {grid_search.best_params_}")
print(f"Test RÂ²: {r2:.4f}")
print(f"Test RMSE: {rmse:.2f}")

# --- 12. Visualization

# 12a. Predicted vs Actual
plt.figure(figsize=(7, 7))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6, edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual Rentals")
plt.ylabel("Predicted Rentals")
plt.title("Predicted vs Actual Rentals")
plt.grid(True)
plt.show()

# 12b. Residual plot
residuals = y_test - y_pred
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_pred, y=residuals, alpha=0.6, edgecolor='k')
plt.axhline(0, color='r', linestyle='--')
plt.xlabel("Predicted Rentals")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Residuals vs Predicted")
plt.grid(True)
plt.show()

# 12c. Correlation heatmap (numeric features)
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()


# 12d. Distribution of target variable
plt.figure(figsize=(8, 6))
sns.histplot(y, bins=30, kde=True)
plt.title("Distribution of Bike Rentals")
plt.xlabel("Number of Rentals")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
